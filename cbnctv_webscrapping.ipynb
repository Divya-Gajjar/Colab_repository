```
!pip install nltk
!pip install spacy
import requests
from bs4 import BeautifulSoup
import pandas as pd 
pd.set_option('display.max_rows', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```
urls = []
results = []
for i in range(1,51):
  url = 'https://www.cnbctv18.com/tags/indigo.html/page-'+str(i)+'/'
  urls.append(url)
  
for i in range(0,50):
  r = requests.get(urls[i])
  soup = BeautifulSoup(r.text,'html.parser')
  for g in soup.find_all('div', class_='article-holder clearfix'):
      anchors = g.find_all('a')
      if anchors:
          link = anchors[0]['href']
          item = {'Link': link}
          
      results.append(item)
results = pd.DataFrame(results)
```

```
results
```

```
def parse(seed_urls):
  df = pd.DataFrame(columns=['x_article_publishon','article_description'])
  for url in seed_urls:
    full_url = requests.get(url)
    page_content = BeautifulSoup(full_url.content, 'html.parser')
    container = page_content.findAll('article',{'class':'article-inner clearfix'})

    for item in container:
      try:
        article_publishon = item.find('div',{'class':'article-wrap'}).text.replace('\n','')
      except:
        article_publishon = None

    container2 = page_content.findAll('div',{'class':'arti-right'})
    for item in container2:
      try:
        article_description = item.find('aside',{'class':'details'}).text.replace('\n','')
      except:
        article_description = None

    df = df.append({'article_description':article_description,'x_article_publishon':article_publishon},ignore_index=True)

  return df
```

```
df_cnbctv = pd.DataFrame(parse(results.Link))
df_cnbctv[['x','article_title','article_publishon']] = df_cnbctv.x_article_publishon.str.split("\r",expand=True)
df_cnbctv = df_cnbctv.drop(['x_article_publishon','x'],axis=1)
print(df_cnbctv)
```
